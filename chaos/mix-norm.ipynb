```python
# plot sin(2m(pi)x) to compare difference

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.sin(2*x)
y5 = np.sin(5*x)
y10 = np.sin(10*x)
y15 = np.sin(15*x)
y20 = np.sin(20*x)

plt.figure(1)

ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,3,4)
ax5 = plt.subplot(2,3,5)
ax6 = plt.subplot(2,3,6)

plt.sca(ax1)
plt.plot(x, y1)
plt.sca(ax2)
plt.plot(x, y2)
plt.sca(ax3)
plt.plot(x, y5)
plt.sca(ax4)
plt.plot(x, y10)
plt.sca(ax5)
plt.plot(x, y15)
plt.sca(ax6)
plt.plot(x, y20)
plt.savefig('sin(2mpix).png')
plt.show()
```
