```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
l1path = "/Users/liliang/Desktop/Optimal-Mixing/simulation/Result/fixed_energy_s/l1norm_histogram.txt"
l2path = "/Users/liliang/Desktop/Optimal-Mixing/simulation/Result/fixed_energy_s/l2norm_histogram.txt"
mixpath = "/Users/liliang/Desktop/Optimal-Mixing/simulation/Result/fixed_energy_s/mixnorm_histogram.txt"
l1 = np.genfromtxt(l1path, dtype=[int, float])
l1pd = pd.DataFrame(l1)
sns.lineplot(data=l1pd,x='f0',y='f1')
plt.show()

l2 = np.genfromtxt(l2path, dtype=[int, float])
l2pd = pd.DataFrame(l2)
sns.lineplot(data=l2pd,x='f0',y='f1')
plt.show()

mix = np.genfromtxt(mixpath, dtype=[int, float])
mixpd = pd.DataFrame(mix)
sns.lineplot(data=mixpd,x='f0',y='f1')
plt.show()
```
